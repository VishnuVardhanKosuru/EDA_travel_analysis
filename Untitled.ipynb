```python
import pandas as pd
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Define function to generate synthetic data
def generate_synthetic_data(num_samples):
    data = []

    for _ in range(num_samples):
        # Arrival and Departure Delay in minutes
        arrival_delay = np.random.randint(0, 61)  # 0 to 60 minutes
        departure_delay = np.random.randint(0, 61)  # 0 to 60 minutes

        # Passenger Type: 0 for Adult, 1 for Teenager, 2 for Senior Citizen
        passenger_type = np.random.choice(['Adult', 'Teenager', 'Senior Citizen'])

        # Ratings on a scale of 1 to 5
        cleanliness = np.random.randint(1, 6)
        onboard_service = np.random.randint(1, 6)
        booking_service = np.random.randint(1, 6)

        # Food and Beverages as 'Yes' or 'No'
        food_and_beverages = np.random.choice(['Yes', 'No'])

        seat_comfort = np.random.randint(1, 6)
        entertainment = np.random.randint(1, 6)

        # Customer Rating
        customer_rating = np.random.randint(1, 6)

        # Append to the data list
        data.append([arrival_delay, departure_delay, passenger_type, cleanliness, onboard_service,
                     booking_service, food_and_beverages, seat_comfort, entertainment, customer_rating])

    # Create a DataFrame
    columns = ['ArrivalDelay', 'DepartureDelay', 'PassengerType', 'Cleanliness', 'OnboardService',
               'BookingService', 'FoodAndBeverages', 'SeatComfort', 'Entertainment', 'CustomerRating']
    df = pd.DataFrame(data, columns=columns)

    return df

# Generate synthetic data with 1200 samples
synthetic_data = generate_synthetic_data(1200)

# Save the synthetic dataset to a CSV file
synthetic_data.to_csv('new_transport_dataset.csv', index=False)
```

```python
def add_missing_values(df, missing_prob=0.05, exclude_columns=None):
    for column in df.columns:
        # Exclude specified columns
        if exclude_columns is not None and column in exclude_columns:
            continue

        # Generate a random mask with the specified probability
        missing_mask = np.random.rand(len(df)) < missing_prob
        # Set values to NaN based on the mask
        df.loc[missing_mask, column] = np.nan
    return df

# Add missing values to df2 excluding 'CustomerRating'
exclude_columns = ['CustomerRating']
df2_with_missing = add_missing_values(df2, missing_prob=0.05, exclude_columns=exclude_columns)
df2_with_missing.head()
```

```python
# Handling missing values for numerical variables
df2['ArrivalDelay'].fillna(df2['ArrivalDelay'].mean(), inplace=True)
df2['DepartureDelay'].fillna(df2['DepartureDelay'].mean(), inplace=True)

# Handling missing values for categorical variables
df2['PassengerType'].fillna(df2['PassengerType'].mode()[0], inplace=True)
df2['Cleanliness'].fillna(df2['Cleanliness'].mean(), inplace=True)
df2['OnboardService'].fillna(df2['OnboardService'].mean(), inplace=True)
df2['BookingService'].fillna(df2['BookingService'].mean(), inplace=True)
df2['FoodAndBeverages'].fillna(df2['FoodAndBeverages'].mode()[0], inplace=True)
df2['SeatComfort'].fillna(df2['SeatComfort'].mean(), inplace=True)
df2['Entertainment'].fillna(df2['Entertainment'].mean(), inplace=True)
df2['CustomerRating'].fillna(df2['CustomerRating'].mean(), inplace=True)

# Check if there are any remaining missing values
df2.isnull().sum()
```

```python
from sklearn.neighbors import LocalOutlierFactor

# Selecting relevant features for LOF (adjust based on your dataset)
features_for_lof = ['ArrivalDelay', 'DepartureDelay', 'Cleanliness', 'OnboardService', 
                    'BookingService', 'SeatComfort', 'Entertainment', 'CustomerRating']

# Creating a subset DataFrame with only the selected features
df2_subset = df2[features_for_lof].copy()

# Handling missing values if any (you can use different strategies based on your needs)
df2_subset.fillna(df2_subset.mean().round(), inplace=True)  # Round the mean to preserve data types

# Initialize the LOF model
lof_model = LocalOutlierFactor(n_neighbors=20, contamination=0.05)

# Fit the model and predict outliers
df2['LOF_Score'] = lof_model.fit_predict(df2_subset)

# Filter out the outliers
df2_no_outliers = df2[df2['LOF_Score'] == 1]

# Check the modified DataFrame without outliers
df2_no_outliers.head()
```
